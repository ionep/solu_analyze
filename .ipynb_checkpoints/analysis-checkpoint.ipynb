```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
```

```python
df=pd.read_excel("col-fixed.xlsx", engine='openpyxl')
```

```python
backupdf=df
backupdf.head()
```

```python
df=backupdf
```

```python
# df = df.iloc[2: , :]
df.head()
df.tail()
```

```python
df.dtypes
```

```python
# df.astype({'TIMESTAMP': 'datetime64'}).dtypes
tempData=df
tempData['Load_R_Power']=pd.to_numeric(tempData["Load_R_Power"])
tempData['TIMESTAMP'] = pd.to_datetime(tempData['TIMESTAMP'])
tempData.head()
```

```python
def monthPlot(data,month,column):
    plotData=data[data.TIMESTAMP.dt.month==month]
    tempData[column] = tempData[column].astype(float)
    fig, ax = plt.subplots()
    fig.set_size_inches(11.7, 8.27)
    ax.grid(False)
    ax.set(xlabel="Date", ylabel=column)

    fig=sns.lineplot(
        data=plotData,
        x="TIMESTAMP", y=column
    )
    plt.savefig('images/'+column+'_month_'+str(month)+'.png')
```

```python
monthPlot(df,2,"Load_R_Power")
monthPlot(df,3,"Load_R_Power")
monthPlot(df,4,"Load_R_Power")
```

```python
def minPlot(data,month,column):
    monthData=df[df.TIMESTAMP.dt.month==month]
    minData=monthData[monthData[column]==monthData[column].min()]
    day=minData.TIMESTAMP.dt.day.values.flatten()[0]
    plotData=monthData[monthData.TIMESTAMP.dt.day==day]
    plotData[column] = plotData[column].astype(float)
    fig, ax = plt.subplots()
    fig.set_size_inches(11.7, 8.27)
    ax.grid(False)
    ax.set(xlabel="Time", ylabel=column)
    ax.xaxis.set_major_formatter(DateFormatter('%H:%M'))

    fig=sns.lineplot(
        data=plotData,
        x="TIMESTAMP", y=column
    )
    plt.savefig('images/'+column+'_month_'+str(month)+'_day_'+str(day)+'_min.png')
```

```python
minPlot(df,2,"Load_R_Power")
minPlot(df,3,"Load_R_Power")
minPlot(df,4,"Load_R_Power")
```

```python
def maxPlot(data,month,column):
    monthData=df[df.TIMESTAMP.dt.month==month]
    maxData=monthData[monthData[column]==monthData[column].max()]
    day=maxData.TIMESTAMP.dt.day.values.flatten()[0]
    plotData=monthData[monthData.TIMESTAMP.dt.day==day]
    plotData[column] = plotData[column].astype(float)
    fig, ax = plt.subplots()
    fig.set_size_inches(11.7, 8.27)
    ax.grid(False)
    ax.set(xlabel="Time", ylabel=column)
    ax.xaxis.set_major_formatter(DateFormatter('%H:%M'))

    fig=sns.lineplot(
        data=plotData,
        x="TIMESTAMP", y=column
    )
    plt.savefig('images/'+column+'_month_'+str(month)+'_day_'+str(day)+'_max.png')
```

```python
maxPlot(df,2,"Load_R_Power")
maxPlot(df,3,"Load_R_Power")
maxPlot(df,4,"Load_R_Power")
```

```python
def plotAll():
    columns=["Load_R_Power","Frequency_Load_R_Phase_Avg","Final Voltage of R Phase","Amperage_Load_R_Phase_Avg","Load_Y_Power","Frequency_Load_Y_Phase_Avg","Load_Y_Voltage","Amperage_Load_Y_Phase_Avg","Load_B_POWER","Frequency_Load_B_Phase_Avg","Real Voltage for plot B-Phase","Amperage_Load_B_Phase_Avg","Real Gen power for Plot R_phase","Frequency_Gen_R_Phase_Avg","Voltage gen Plot for R-Phase","Amperage_Gen_R_Phase_Avg","Real Power gen for plot Y-Phase","Frequency_Gen_Y_phase_Avg","Voltage Gen for plot Y Phase","Amperage_Gen_Y_phase_Avg","Real power Gen for plot B phase","Frequency_Gen_B_Phase_Avg","Voltage Gen for Plot B- Phase","Amperage_Gen_B_Phase_Avg",""]
    for i in range(2,5):
        for j in columns:
            print(j)
            monthPlot(df,i,j)
            minPlot(df,i,j)
            maxPlot(df,i,j)
```

```python
plotAll()
```
